```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter

sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"C:\Users\brahm\Downloads\zomato.csv", encoding="latin-1")

# First look at the data
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
#basic info about the data
print(df.shape)
print(df.columns)
df.info()
```

```python
# Rate cleaning
df["rate_clean"] = df["rate"].astype(str)
df["rate_clean"] = df["rate_clean"].replace(["NEW", "-", "nan"], np.nan)
df["rate_clean"] = df["rate_clean"].str.replace("/5", "", regex=False)
df["rate_clean"] = pd.to_numeric(df["rate_clean"], errors="coerce")
print("DONE")
```

```python
# Cost cleaning
df["cost"] = df["approx_cost(for two people)"].astype(str)
df["cost"] = df["cost"].str.replace(",", "")
df["cost"] = pd.to_numeric(df["cost"], errors="coerce")
```

```python
# drop duplicates
df = df.drop_duplicates()
```

```python
#Remove rows without rating
df = df[df["rate_clean"].notna()]
```

```python
print(df.describe(include='all').T)
print("\nMissing values:\n", df.isna().sum())
```

```python
# Rating distribution
plt.figure(figsize=(8,5))
sns.histplot(df["rate_clean"].dropna(), bins=20, kde=True)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.show()
```

```python
df["votes"] = pd.to_numeric(df["votes"], errors="coerce")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["votes"].dropna(), bins=50, color="orange")
plt.title("Votes Distribution")
plt.xlabel("Votes")
plt.show()
```

```python
# Cost distribution
plt.figure(figsize=(8,5))
sns.histplot(df["cost"].dropna(), bins=20, kde=True, color="green")
plt.title("Cost for Two Distribution")
plt.xlabel("Cost")
plt.show()
```

```python
# Online order Vs Rating
sns.boxplot(data=df, x="online_order", y="rate_clean")
plt.title("Online Order vs Rating")
plt.show()
```

```python
# Top locations
plt.figure(figsize=(12,8))

loc_ratings = df.groupby("location")["rate_clean"].mean().sort_values(ascending=False)[:20]

sns.barplot(x=loc_ratings.values, y=loc_ratings.index)
plt.title("Top 20 Locations by Avg Rating")
plt.xlabel("Average Rating")
plt.show()
```

```python
# Top 10 Locations with Most Restaurants
plt.figure(figsize=(12, 6))
top_locations = df['location'].value_counts().head(10)
sns.barplot(x=top_locations.values, y=top_locations.index, palette='viridis')
plt.title('Top 10 Locations with Most Restaurants in Bangalore', fontsize=16, fontweight='bold')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

```python
#Top Cuisines
cuisine_counter = Counter(", ".join(df["cuisines"].dropna()).split(", "))
top_cuisines = pd.DataFrame(cuisine_counter.most_common(15), columns=["Cuisine","Count"])

sns.barplot(data=top_cuisines, x="Count", y="Cuisine")
plt.title("Top 15 Popular Cuisines")
plt.show()
```

```python
# restaruant types
rest_counter = Counter(", ".join(df["rest_type"].dropna()).split(", "))
rt = pd.DataFrame(rest_counter.most_common(10), columns=["Rest Type", "Count"])

sns.barplot(data=rt, x="Count", y="Rest Type")
plt.title("Top 10 Restaurant Types")
plt.show()
```

```python
# Top 10 Locations with Most Restaurants
plt.figure(figsize=(12, 6))
top_locations = df['location'].value_counts().head(10)
sns.barplot(x=top_locations.values, y=top_locations.index, palette='viridis')
plt.title('Top 10 Locations with Most Restaurants in Bangalore', fontsize=16, fontweight='bold')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

```python
# Corelation heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df[["rate_clean", "votes", "cost"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Key Insights Summary
print("=" * 55)
print("   ZOMATO BANGALORE RESTAURANTS - KEY INSIGHTS")
print("=" * 55)

print(f"\n📊 Total Restaurants Analyzed: {df.shape[0]:,}")

print(f"\n📍 Top Location: {df['location'].value_counts().index[0]}")
print(f"   Restaurants: {df['location'].value_counts().values[0]:,}")

# ✅ FIXED: use rate_clean
print(f"\n⭐ Average Rating: {df['rate_clean'].mean():.2f} / 5")
print(f"   Highest Rated Area: {df.groupby('location')['rate_clean'].mean().idxmax()}")

# ✅ FIXED: use cleaned cost column
print(f"\n💰 Average Cost for Two: ₹{df['cost'].mean():.0f}")
print(f"   Most Expensive Area: {df.groupby('location')['cost'].mean().idxmax()}")

print(f"\n🍽️ Most Popular Cuisine: {df['cuisines'].value_counts().index[0]}")

# safer handling (in case Yes/No missing)
online_counts = df['online_order'].value_counts()

print(f"\n📱 Restaurants with Online Order: {online_counts.get('Yes', 0):,}")
print(f"   Restaurants without Online Order: {online_counts.get('No', 0):,}")

print("\n" + "=" * 55)
print("✅ Analysis Complete!")
print("=" * 55)
```
